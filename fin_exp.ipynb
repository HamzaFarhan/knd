```python
import datetime
import random

import pandas as pd
```

```python
cols = ["date", "description", "category", "amount"]

# Generate sample financial data for a software company


# Sample descriptions and categories
descriptions = [
    "PAYMENT - ADOBE CREATIVE CLOUD",
    "AWS SERVICES DEBIT",
    "GITHUB SUBSCRIPTION",
    "STRIPE PAYMENT PROCESSING FEE",
    "AMAZON.COM PURCHASE - COMPUTER EQUIPMENT",
    "ACH TRANSFER - FREELANCE DEVELOPER",
    "DEPOSIT - CLIENT PAYMENT",
    "UBER EATS - TEAM LUNCH",
    "EVENTBRITE - TECH CONFERENCE TICKETS",
    "ZOOM VIDEO COMM SUBSCRIPTION",
    "INVOICE PAYMENT - CLIENT A",
    "QUARTERLY BONUS RECEIVED",
    "TAX REFUND DEPOSIT",
    "CONSULTING REVENUE - PROJECT X",
    "INVESTMENT DIVIDEND",
]
categories = ["Software", "Infrastructure", "Development", "Hardware", "Revenue", "Operations"]

# Define which descriptions are income (money received)
income = [
    "DEPOSIT - CLIENT PAYMENT",
    "INVOICE PAYMENT - CLIENT A",
    "QUARTERLY BONUS RECEIVED",
    "TAX REFUND DEPOSIT",
    "CONSULTING REVENUE - PROJECT X",
    "INVESTMENT DIVIDEND",
]

# Create sample data
data = []
for i in range(10):
    date = datetime.datetime.now() - datetime.timedelta(days=random.randint(1, 30))
    description = random.choice(descriptions)
    category = random.choice(categories)
    amount = round(random.uniform(100.0, 5000.0), 2)
    # Make amount negative if it's not in income list
    if description not in income:
        amount = -amount
    data.append([date, description, category, amount])
```

```python
df = pd.DataFrame(data, columns=cols)
df
```

```python
# Create a grocery store inventory DataFrame
grocery_items = ["Apples", "Bread", "Milk", "Eggs", "Bananas", "Chicken", "Rice", "Pasta", "Tomatoes", "Cereal"]

# Set half of the items to have 0 stock
stocks = []
for _ in range(5):
    stocks.append(0)
for _ in range(5):
    stocks.append(random.randint(5, 50))

# Shuffle to randomize which items have 0 stock
random.shuffle(stocks)

# Create the DataFrame
grocery_df = pd.DataFrame({"item": grocery_items, "stock": stocks})

grocery_df
```

```python
in_stock_items = grocery_df[grocery_df["stock"] > 0]
in_stock_items
```

```python
out_of_stock_items = grocery_df[grocery_df["stock"] == 0]
out_of_stock_items
```

```python
# Create a DataFrame with a date column in datetime format
import random
from datetime import datetime, timedelta

import pandas as pd

# Create base date (today)
base_date = datetime.now()

# Generate 10 dates with varying months
dates = []
for i in range(10):
    # Vary the months by subtracting a random number of days (0-90 days)
    # This will create dates spanning approximately 3 months
    random_days = random.randint(0, 90)
    dates.append(base_date - timedelta(days=random_days))

# Create DataFrame with just the date column
date_df = pd.DataFrame({"date": dates})

date_df
```

```python
date_df["day"] = date_df["date"].dt.day
date_df["month"] = date_df["date"].dt.month
date_df["year"] = date_df["date"].dt.year

date_df
```
